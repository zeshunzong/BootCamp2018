```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**4 - 14 * x**3 + 60 * x**2 - 70*x

def f_prime(x):
    return 4 * x**3 - 42 * x**2 + 120 * x - 70

def secant_find_min(f_prime, initGuess1, initGuess2, epsilon):
    difference = epsilon + 1
    x_km1 = initGuess1
    x_k = initGuess2
    maxItr = 10000
    currentItr = 0
    
    #plot f 
    #plt.figure(figsize = (20,20))
    plt.plot(np.linspace(0.5, 1.0, 50), f(np.linspace(0.5, 1.0, 50)), label = r"$f(x)$")
    
    while (currentItr < maxItr) and (difference > epsilon):
        #print((f_prime(x_k) - f_prime(x_km1)))
        x_kp1 = x_k - f_prime(x_k) * (x_k - x_km1) / (f_prime(x_k) - f_prime(x_km1))
        difference = abs(x_kp1 - x_k)
        currentItr += 1
        x_km1 = x_k
        x_k = x_kp1
        plt.plot(x_k, f(x_k), marker = "o", label = str(currentItr))
        
    plt.legend(loc = "upper right")
    plt.title("Find local min")
    plt.show()
    if (currentItr == maxItr):
        raise ValueError("Does not converge")
    return x_k

secant_find_min(f_prime, 0, 0.1, 1e-12)
```
